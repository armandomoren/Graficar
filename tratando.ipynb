```python
t = np.linspace(0,50,100)
plt.figure(figsize = (7,4))
plt.plot(t, .5*np.cos(.5 *t) + .1 * np.sin(.5 *t), '-', lw = 1, ms = 4,label = '$x(t)$' )
plt.plot(t, .5*.1*np.cos(.5 *t) - .5*.5 * np.sin(.5*t), 'bo-', lw =1, ms = 7,label = r'$\dot{x(t)}$')
plt.xlabel('$t$', fontsize = 20)
plt.show()



```
